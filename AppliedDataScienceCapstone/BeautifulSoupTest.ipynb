```python
mainurl = "https://www.hurriyetemlak.com/istanbul-satilik?counties=bakirkoy,besiktas,kadikoy"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36"}
mainrequest = requests.get(mainurl, headers)
print(mainrequest.status_code)
mainsoup = BeautifulSoup(mainrequest.text,'html.parser', )
pagestart = mainsoup.select("section.pagination li.page-item")[0].text
pageend   = mainsoup.select("section.pagination li.page-item")[-1].text
```

# For Multiple Pages

```python
%%time
start = time.process_time()
data = []
for pagenum in range(int(pagestart), int(pageend)+1):
    currentpage = mainurl+"&page="+str(pagenum)
    print("******************************************************************************************************************")
    print("PageNo: ", pagenum, "; URL: ", currentpage)
    if pagenum == 1:
        currentsoup = mainsoup
        print("STATUS: ", mainrequest.status_code)
    else: 
        currentrequest = requests.get(currentpage, headers)
        print("STATUS: ", currentrequest.status_code)
        currentsoup = BeautifulSoup(currentrequest.text,'html.parser')
    
    for i in currentsoup.select("div.listing-item"):
        
        if not i.select("div.listingView.projeland-adv.type-1000010100"):
            
            row = []
            if i.select("div.upper.sibling div.top div.list-view-price"):
                row.append(i.select("div.upper.sibling div.top div.list-view-price")[0].text)
            else:
                row.append("Null")
            if i.select("div.upper.sibling div.top div.list-view-date"):
                row.append(i.select("div.upper.sibling div.top div.list-view-date")[0].text)
            else:
                row.append("Null")
            if i.select("div.middle.sibling div.top div.left"):
                row.append(i.select("div.middle.sibling div.top div.left")[0].text)
            else: 
                row.append("Null")
            if i.select("div.middle.sibling div.top div.right span.celly.houseRoomCount"):
                row.append(i.select("div.middle.sibling div.top div.right span.celly.houseRoomCount")[0].text)
            else:
                row.append("Null")
            if i.select("div.middle.sibling div.top div.right span.celly.squareMeter.list-view-size"):
                row.append(i.select("div.middle.sibling div.top div.right span.celly.squareMeter.list-view-size")[0].text)
            else: 
                row.append("Null")
            if i.select("div.middle.sibling div.top div.right span.celly.buildingAge"):
                row.append(i.select("div.middle.sibling div.top div.right span.celly.buildingAge")[0].text)
            else:
                row.append("Null")
            if i.select("div.middle.sibling div.top div.right span.celly.floortype"):
                row.append(i.select("div.middle.sibling div.top div.right span.celly.floortype")[0].text)
            else: 
                row.append("Null")
            if i.select("div.bottom.sibling div.left div.upper div.list-view-header"):
                row.append(i.select("div.bottom.sibling div.left div.upper div.list-view-header")[0].text)
            else:
                row.append("Null")
            if i.select("div.bottom.sibling div.left div.down div.list-view-location"):
                row.append(i.select("div.bottom.sibling div.left div.down div.list-view-location")[0].text)
            else:
                row.append("Null")
            
            data.append(row)
    
    print("DURATION: ", time.process_time() - start)
```

```python
pd.DataFrame(data, columns = ["price" ,"date", "type", "roomcount", "area", "areaunit", "age", "floortype", "location"])
```

```python
data = []
for i in mainsoup.select("div.list-view-content"):
    row = []
    if i.select("div.upper.sibling div.top div.list-view-price"):
        price = i.select("div.upper.sibling div.top div.list-view-price")[0].text.replace(",","")
        currency = price.split()[1]
        price = price.split()[0]
        row.append(price)
        row.append(currency)
    if i.select("div.upper.sibling div.top div.list-view-date"):
        date = i.select("div.upper.sibling div.top div.list-view-date")[0].text
    if i.select("div.middle.sibling div.top div.left"):
        type = i.select("div.middle.sibling div.top div.left")[0].text
    if i.select("div.middle.sibling div.top div.right span.celly.houseRoomCount"):
        roomcount = i.select("div.middle.sibling div.top div.right span.celly.houseRoomCount")[0].text
    if i.select("div.middle.sibling div.top div.right span.celly.squareMeter.list-view-size"):
        squaremeter = i.select("div.middle.sibling div.top div.right span.celly.squareMeter.list-view-size")[0].text
        area = squaremeter.split()[0]
        areaunit = squaremeter.split()[1]
    if i.select("div.middle.sibling div.top div.right span.celly.buildingAge"):
        age = i.select("div.middle.sibling div.top div.right span.celly.buildingAge")[0].text
    if i.select("div.middle.sibling div.top div.right span.celly.floortype"):
        floortype =i.select("div.middle.sibling div.top div.right span.celly.floortype")[0].text
    if i.select("div.bottom.sibling div.left div.upper div.list-view-header"):
        desc = i.select("div.bottom.sibling div.left div.upper div.list-view-header")[0].text
    if i.select("div.bottom.sibling div.left div.down div.list-view-location"):
        location = i.select("div.bottom.sibling div.left div.down div.list-view-location")[0].text
    
    list = [price, currency ,date, type, roomcount, area, areaunit, age, floortype, location]
    row = [' '.join(i.replace("\n", "").strip().split()) for i in list]
    data.append(row)
```
